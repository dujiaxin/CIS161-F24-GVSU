```python
# a simple repeat baseline
df = pd.read_csv('cityC_challengedata.csv.gz', compression='gzip')

# Create a new dataframe with all possible combinations of uid, d, and t
uids = df['uid'].unique()
days = range(75)
times = range(48)  # 0 to 47

all_combinations = pd.MultiIndex.from_product([uids, days, times], names=['uid', 'd', 't']).to_frame(index=False)
merged_df = all_combinations.merge(df, on=['uid', 'd', 't'], how='left')

# Forward fill missing values for each user and day
merged_df.sort_values(by=['uid', 'd', 't'], inplace=True)
merged_df[['x', 'y']] = merged_df.groupby(['uid', 'd'])[['x', 'y']].ffill()

# This is your new dataframe with missing values filled
merged_df=merged_df.ffill()
# Filter data for days 46 to 59
previous_two_weeks = merged_df[merged_df['d'].between(46, 60)]

# Create a copy of the data and adjust the day values to 60 to 74
predicted_data = previous_two_weeks.copy()
predicted_data['d'] = predicted_data['d'] + 14
result_df = df.merge(predicted_data, on=['uid', 'd', 't'], how='inner',suffixes=('_real', '_predicted'))
# save the results
result_df = result_df[result_df['x_real'] == 999]
result_df = result_df.rename(columns={"x_predicted": "x", "y_predicted": "y"})
# Replace 999 with NaN for easier handling
result_df['x'] = result_df['x'].where(result_df['x'] != 999, pd.NA)
result_df['y'] = result_df['y'].where(result_df['y'] != 999, pd.NA)

# Use the 'ffill' method to forward fill the NaN values with the previous non-NaN value
result_df = result_df.fillna(method='ffill')
result_df[["uid",	"d",	"t",	"x",	"y"]].astype(int).to_csv('jiaxin_cityC_humob24.csv.gz',compression='gzip',index=False)
```
